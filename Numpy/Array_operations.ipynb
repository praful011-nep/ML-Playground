**Deep and Shallow Copy**  
*Shallow Copy: When a arry is sliced  and performed operations in it sliced copy, no changes will occur in the original array as both are in seperate location in RAM*  
*Deep Copy: When a array is sliced and performed operations in it, the original array also changes as both variables are in the same location of RAM*

```python
a=np.arange(1,21)
a
```

```python
slice = a[:5]
slice = slice *10
slice
```

```python
a
```

```python
a[4:6] *10 
```

```python
a
```

```python
b = a.copy() #Shallow Copy
# b=a     #Deep Copy
b[0] = 99 
```

```python
b
```

```python
a
```

**Advanced Array Manipulation**

Stacking Array  
*Can't stack 2 matrix  but can stack or merge 2 vectors. The stacked vectors are stacked in the form of matrix or vectors vertically (row wise) or horizontally (column wise)*  
np.vstack((*Matrix1, Matrix2*)) [Vertically]  *The stacked vectors are stored in the form of matrix vertically (row wise)*  
np.hstack((*Matrix1, Matrix2*))  [Horizontally] *The stacked arrays are stored in vector form merged Horizontally (column wise)*  
np.column_stack((*Matrix1, Matrix2*)) [Horizontally] *The stacked arrays are stored in matrix forms Horizontally(Column wise)*

```python
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
```

```python
np.vstack((a,b)) #Vertically
```

```python
np.hstack((a,b)) #Horizontally
```

```python
np.column_stack((a,b))
```

Splitting Arrays  
*Matrix can be split in to number of parts*  
np.split(*Matrix*, **) and np.vsplit() ; splits the array row wise  
np.hsplit() ; splits the array column wise

```python
c = np.arange(16).reshape(4, 4)
c
```

```python
np.split(c,4)
```

```python
np.hsplit(c,4)
```

```python
np.vsplit(c,4) 
```
