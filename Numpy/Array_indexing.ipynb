```python
import numpy as np
```

**Indexing and Slicing of Vectors**

```python
arr=np.arange(11,21)
arr
```

Indexing

```python
#Indexing works just like lists (starts from 0)
arr[2]
```

Slicing

```python
#name[start:end] (start - include, end - exclude)
arr[:3] #if startign value is not given the default is 0 and if end is not given array is sliced till last element
```

```python
arr[:] #whole array is returend if not start or end index are given
```

Steps is Slicing

```python
#name[start::steps]
arr[3::2]
```

**Indexing and Slicing of Matrix**

```python
a=np.arange(1,31).reshape(6,5)
a
```

Indexing

```python
# name[row,column]
a[5,2]
```

Slicing

```python
#THe sliced portion is returend in vector if only one dimension is given
a
```

```python
#If only one index is given then that will target rows
a[1]
```

```python
#For a portion, slice rows as well as columns
a[:2,2:4]
```

```python
a[3:,3:]
```

```python
# Slicing a entire column
a[:,2]
```

****Boolean Indexing****

Uses boolean mask i.e. True and False, to filter and extract specific elements(True) from a array


```python
ar=np.arange(11,21)
ar
```

```python
#Returns the boolean output of every index of ar creating a new list
bool_index=ar%2==0
bool_index
```

```python
#Extracts only the True output indices.
ar=ar[bool_index]
ar
```
