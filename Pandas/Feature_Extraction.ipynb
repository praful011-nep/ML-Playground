*Filtering no of episdoes from title*  
As enclosed inside "()"

```python
def extract_episodes(txt):
    check=False
    data = ""
    for i in txt:
        if i== ")":
            check = False
            return data
        if check == True:
            data += i
        if i=="(":
            check=True
          
```

```python
#Applying the funciton
df["Episodes"] = df["Title"].apply(extract_episodes)
```

Remove " eps" from Episodes  
Type conversion from str to int

```python

df["Episodes"]=df["Episodes"].str.replace(" eps","").astype(int)
```
