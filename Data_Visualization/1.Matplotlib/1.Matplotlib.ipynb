**Subplots in Matplotlib**

Multiple plots at same time

plt.subplot(*no of rows, no of columns, selected subplot*)

*To plot in any of the subplot*  
Just use plot function just below the respective subplot funciton

Subplots cannot be selected normally

```python
plt.subplot(2,2,1)
plt.plot(b,a)
plt.subplot(2,2,2)
plt.plot(a,b)
plt.subplot(2,2,3)
m = plt.plot(a,b-2)
plt.subplot(2,2,4)
plt.plot(b+5,a**2)
```

**Object Oriented Way**

To select subplots, objected oriented way is used

As the canvas works as an image. So, the subplots cannot change positions  
i.e. *Have full control over canvas*

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

<pre>
<b>.figure:</b>Creates the canvas
.figure(figsize=<i>(width,height)</i>,dpi=<i>how much to pixelate</i>)
The attributes are optional. In default figsize=(6.4,4.8) and dpi = 100
<b>.add_axes:</b> it specifies size and position of the object
.add_axes([
    <i>move along x-axis<i>,
    <i>move along y-axis<i>,
    <i>width<i>,
    <i>height<i>
])
</pre>

```python
fig = plt.figure(figsize=(12,8), dpi=100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(a,b)
axis2 = fig.add_axes([0.2,0.55,.5,.5])
axis2.plot(a,b)
```
