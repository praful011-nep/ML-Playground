```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
import seaborn as sns
```

****Categorial Plot****  
Used to visualize qualitative data.  
*Different plots for categorical data:*
* boxplot
* voilinplot
* stripplot
* swarmplot
* barplot
* countplot


```python
df = sns.load_dataset("tips")
df
```

**Count Plot**  
* Counts the categories of a column
* Only takes 1 column at once
* sns.countplot(*column_to_categorizw*, hue = *(if want data further categorized on the basis of another column)*)

```python
# How many male and female smoke?
sns.countplot(x = df['sex'], hue = df['smoker'])
```

**Bar Plot** 
* Displays a statistical summary (such as the mean, median, or another estimator) of a numerical variable for each category.
* One axis holds qualitative data and another holds quantitative data
* sns.barplot( x = *categorical_column*, y = *distributive_column*, estimator = *specifies statistical summary needed*)
* estimator is set to mean in defualt
* The line on the top of each bar is called *Error Bar*
* Error Bar is a vertical line drawn on top of each rectangular bar to represent the statistical uncertainty or variability of the underlying data
* By default it shows 95% confident interval.
* Can be changed through errorbar parameter.  

| Objective | Argument `errorbar=` | What It Displays |
| :--- | :--- | :--- |
| **Statistical Certainty** | `('ci', 95)` *(Default)* | The 95% confidence interval of the estimated mean. |
| **Data Spread** | `'sd'` | The standard deviation, showing the average distance of points from the mean. |
| **Data Percentiles** | `'pi'` | A percentile interval (e.g., showing where 95% of the data falls). |
| **Remove Bars** | `None` | Disables error bars entirely for a cleaner look if variability isn't needed. |


```python
# Estimator is mean in default
# Shows graph of mean of total bills on the basis of gender.
sns.barplot(
    x = df['sex'],
    y = df['total_bill']
)
```

Use of estimator

```python
# Plot of Standard Deviation of total bill on the basis of gender
sns.barplot(
    x = df['sex'],
    y = df['total_bill'],
    estimator=np.std
)
```

```python
# Sum of total bills of each sex.
sns.barplot(
    x = df['sex'],
    y = df['total_bill'],
    estimator=np.sum
)
```

**Box Plot**

* Shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable
* sns.bolxplot( x = *quantitative*, y = *qualitative*, data = *dataframe* )

```python
sns.boxplot(x = df['total_bill'], y = df['sex'])
```

```python
sns.boxplot(
    x = 'tip',
    y = 'sex',
    data = df
)
```

**Voilin Plot**

* Combines a box plot and a kernel density plot to show the probability density of numerical data.
* sns.voilinplot( x = *quantitative* , y = *qualitative*, data = *dataframe*)

```python
sns.violinplot(
    x = 'day',
    y = 'tip',
    data = df
)
```

**Strip Plot**

* Displays the distribution of a numerical variable across categories by plotting each individual observation.
* sns.stripplot( x = *qualitative*, y = *quantitative*, data = *dataframe*, jitter = *boolean*)
* jitter = True ; helps reduce overlap

```python
sns.stripplot(
    x = 'day',
    y = 'total_bill',
    data = df
)
```

```python
sns.stripplot(
    x = 'day',
    y = 'total_bill',
    data = df,
    hue = 'sex',
    jitter = True
)
```

**Swarm Plot**

* Displays every individual observation across categories while arranging the points so that they do not overlap.
* Works similar to Voilinplot
* sns.swarnplot( x = *qualitative*, y = *quantitative*, data = *dataframe*)

```python
sns.swarmplot(
    x = 'day',
    y = 'total_bill',
    data = df
)
```

**Combining VoilinPlot and SwarmPlot**

* When both plots are created in the same cell then both get created in the same canvas.

```python
plot = sns.violinplot(
    x = 'tip',
    y = 'day',
    data = df
)
sns.swarmplot(
    x = 'tip',
    y = 'day',
    data = df
)
plot.set_title('Combining Voilin Plot and Swarm Plot')
```
