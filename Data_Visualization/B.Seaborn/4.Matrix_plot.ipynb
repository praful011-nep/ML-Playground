```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

-  Matrix plots are visualizations that represent data in a matrix (grid) format, where colors indicate the magnitude of values.
- Used to visualize **2-dimensional data** effectively.
- Color intensity represents the **value or magnitude** of each cell in the matrix.
- Helps identify **patterns, trends, correlations, and clusters** in data.
- Commonly used during **Exploratory Data Analysis (EDA)**.
- Best suited for **correlation matrices, pivot tables, and contingency tables**.
- Makes it easier to analyze large datasets compared to reading numerical tables.
- A **color bar (legend)** is often included to explain the meaning of different colors.
- The two main matrix plots in Seaborn are:
  - **`sns.heatmap()`** – Visualizes matrix values using colors.
  - **`sns.clustermap()`** – Displays a heatmap with hierarchical clustering of rows and columns.

**Heat Map**

* Visualizes data in a matrix format using colors to represent values.
* Used to identify patterns, correlations, trends, and missing values.
* Best used with correlation matrices, pivot tables, or 2D arrays.
* Darker/lighter colors indicate higher or lower values (depending on the colormap).
* sns.heatmap(*data*, annot=*bool*, cmap=" *color_scheme* ")

* *annot=True*- Displays the values inside each cell.
* *cmap="coolwarm"* - Changes the color scheme.
* *linewidths* - Adds lines between cells.
* *fmt* - Formats the annotation values (e.g., `".2f"`).


```python
tipscorr = tips[['total_bill', 'tip', 'size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr())
```

```python
sns.heatmap(
    tipscorr.corr(),
    annot= True,
    linewidths=2,
    cmap="coolwarm",
)
```

**Cluster Map**

* Similar to a heatmap but also groups similar rows and columns using hierarchical clustering.
* Useful for identifying clusters and similarities in data.
* Best used with correlation matrices or standardized datasets.
* Automatically reorders rows and columns based on similarity.



sns.clustermap(*data*, annot=*bool*, cmap=" *color_scheme* ")

```python
sns.clustermap(tipscorr.corr())
```
