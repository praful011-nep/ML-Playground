**Count Plot**  
* Counts the categories of a column
* Only takes 1 column at once
* sns.countplot(*column_to_categorizw*, hue = *(if want data further categorized on the basis of another column)*)

```python
# How many male and female smoke?
sns.countplot(x = df['sex'], hue = df['smoker'])
```
