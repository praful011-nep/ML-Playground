```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Distribution Plot**  
Used to visualize quantitative data.

Different Plots to Visualize Distribution of dataset:
* distplot/histplot
* jointplot
* pairplot
* rugplot

```python
df = sns.load_dataset('tips')
df
```

First identify quantitative and qualitative data:  
*Quantitative:*  
* total_bill
* tip  
*Qualitative( Categorical):*  
* sex
* smoker
* day  
* time  
* size (as only 6 unique values are present in large)

1. **Distplot/ Histplot**  
sns.displot/histplot(*column*, bins = *range of data taken at once*, kde = *boolean, kernel density estimator*)

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins = 20, kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins = 20, kde=True)
```
