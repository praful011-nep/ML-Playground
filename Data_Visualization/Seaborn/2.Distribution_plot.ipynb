```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Distribution Plot**  
Used to visualize quantitative data.

Different Plots to Visualize Distribution of dataset:
* distplot/histplot
* jointplot
* pairplot
* rugplot

.load_dataset("*dataset_name*")
* To fetch data sets are provided by seaborn itself

```python
df = sns.load_dataset('tips')
df
```

First identify quantitative and qualitative data:  
*Quantitative:*  
* total_bill
* tip  
*Qualitative( Categorical):*  
* sex
* smoker
* day  
* time  
* size (as only 6 unique values are present in large)

1. **Distplot/ Histplot**   
* Used to make a single histogram
* sns.displot/histplot(*column*, bins = *range of data taken at once*, kde = *boolean, kernel density estimation*)

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins = 20, kde=True,)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins = 20, kde=True)
```

**Joint Plot**  
* Used to compare 2 plots  
* sns.jointplot( x = "*col_name*", y = "*col_name*", data = *data_frame*, kind = *type_of_plot (default = scatter)* )
* *Literals in kind*: kind = ["scatter", "hist", "kde", "hex", "reg", "resid"]


```python
sns.jointplot(
    x = "total_bill",
    y = "tip",
    data = df,
    kind = 'scatter'
)
```

```python
sns.jointplot(
    x = "total_bill",
    y = "tip",
    data = df,
    kind = 'hist'
)
```

```python
sns.jointplot(
    x = "total_bill",
    y = "tip",
    data = df,
    kind = 'reg'
)
```
