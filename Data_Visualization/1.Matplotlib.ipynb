****Matplotlib****

Plotting library in Python for creating interactive, animated and static visualization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Matplotlib Basics**

```python
a = np.linspace(0,5,11)
a
```

```python
b = a **2
b
```

<pre>
<i><b>Making a plot</i></b>  
    plt.plot(<i>x-axis, y-axis</i>)  
<i><b>Adding labels</i></b> 
    plt.title("<i>title</i>")  
    plt.xlabel("")  
    plt.ylabel("")
</pre>

**Note:** 
All the labels should be in the same cell as the plot funciton

```python
plt.plot(
    a,
    b
)
plt.title("First plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

**Subplots in Matplotlib**

Multiple plots at same time

plt.subplot(*no of rows, no of columns, selected subplot*)

*To plot in any of the subplot*  
Just use plot function just below the respective subplot funciton

```python
plt.subplot(2,2,1)
plt.plot(b,a)
plt.subplot(2,2,2)
plt.plot(a,b)
plt.subplot(2,2,3)
m = plt.plot(a,b-2)
plt.subplot(2,2,4)
plt.plot(b+5,a**2)
```
