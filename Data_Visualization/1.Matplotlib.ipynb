****Matplotlib****

Plotting library in Python for creating interactive, animated and static visualization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Matplotlib Basics**

```python
a = np.linspace(0,5,11)
a
```

```python
b = a **2
b
```

<pre>
<i><b>Making a plot</i></b>  
    plt.plot(<i>x-axis, y-axis</i>)  
<i><b>Adding labels</i></b> 
    plt.title("<i>title</i>")  
    plt.xlabel("")

**Note:** 
All the labels should be in the same cell as the plot funciton

```python
plt.plot(
    a,
    b
)
plt.title("First plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```
